```python
from collections import defaultdict
import numpy as np
import pandas as pd
import gzip
import json
```

```python
# train = []
with gzip.open("./data/train.json.gz", 'rt', encoding='utf-8') as gzip_file:
    for line in gzip_file:
        train = json.loads(line)
```

```python
# for each user, get all the asin, and the time they bought it
user_asin = defaultdict(list)
for i in range(len(train)):
    user_asin[train[i]['reviewerID']].append((train[i]['asin'], train[i]['unixReviewTime']))
```

```python
user_asin
```

```python
# make a histogram of the number of items bought by all user
hist = defaultdict(int)
for user, asin in user_asin.items():
    hist[len(asin)] += 1

hist
```

```python
# draw a barplot of the histogram
import matplotlib.pyplot as plt
plt.bar(hist.keys(), hist.values())
plt.xlabel('Number of items bought')
plt.ylabel('Number of users')
plt.show()
```

```python
# zoom into the 0-50 range
plt.bar(hist.keys(), hist.values())
plt.xlabel('Number of items bought')
plt.ylabel('Number of users')
plt.xlim(0,50)
plt.show()
```

```python
# get the distribution between the first and last purchase
# for each user, get the first and last purchase
first_last = defaultdict(list)
for user, asin in user_asin.items():
    first_last[user].append(asin[0][1])
    first_last[user].append(asin[-1][1])

first_last
```

```python
# plot the time difference between the first and last purchase
diff = []
for user, times in first_last.items():
    diff.append(times[1] - times[0])


plt.hist(diff, bins=100)
```

```python
# get all the item user pair with vote and without vote
item_user_vote = defaultdict(list)
for i in range(len(train)):
    # check if there is a vote attribute in the dictionary
    if 'vote' in train[i]:
        item_user_vote[train[i]['asin']].append((train[i]['reviewerID'], train[i]['vote']))
```

```python
# get the ratings of all the items
item_rating = defaultdict(list)
for i in range(len(train)):
    item_rating[train[i]['asin']].append(train[i]['overall'])
```

```python
# get number of times that each item was bought
item_count = defaultdict(int)
for i in range(len(train)):
    item_count[train[i]['asin']] += 1
```

```python
# mode number of times that all items were bought
from statistics import mode
mode(item_count.values())

# number of items that were bought only one time
count = 0
for item, num in item_count.items():
    if num == 1:
        count += 1
count
```
